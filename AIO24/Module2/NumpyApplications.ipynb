#Use numpy to convert Image Black and White



```python
import numpy as np
import matplotlib.image as mpimg

!gdown 1ZF09wOx9IxUEuZ2iFdQHOp3udYHCm1-Y
img = mpimg.imread ('/content/ChiPu.jpg')
img
```

##Lightness Method

```python
def lightness(vector):
  return (max(vector) + min(vector))/2

gray_img_01 = np.apply_along_axis(lightness, axis = 2, arr = img)
gray_img_01 = np.clip(gray_img_01, 0, 255).astype(np.uint8)

gray_img_01
```

##Average Method

```python
gray_img_02 = np.apply_along_axis(np.average, axis = 2, arr = img)
gray_img_02 = np.clip(gray_img_02, 0, 255).astype(np.uint8)

gray_img_02
```

##Luminosity Method

```python
def luminosity(vector):
  return vector[0] * 0.21 + vector[1] * 0.72 + vector[2] * 0.07

gray_img_03 = np.apply_along_axis(luminosity, axis = 2, arr = img)
gray_img_03 = np.clip(gray_img_03, 0, 255).astype(np.uint8)

gray_img_03
```

#Use Numpy for data analysis

##Load data

```python
!gdown 1iA0WmVfW88HyJvTBSQDI5vesf-pgKabq
import pandas as pd

df = pd.read_csv ('/content/advertising.csv')

data = df.to_numpy()
```

##Get the index and value of the max sales

```python
max_sales_index = np.argmax(data[:,3])
print(f"Index: {max_sales_index}, Value: {data[max_sales_index,3]}")
```

##Get the mean of TV

```python
mean_TV = np.average(data[:,0])
print(f"Mean TV: {mean_TV}")
```

##Get the number of sales that is >= 20

```python
count_sales = np.sum(data[:,3] >= 20)
print(f"The number of sales that is larger or equal to 20 is {count_sales}")
```

## Get the average Radio where the corresponding Sales is larger or equal to 15

```python
average_radio = np.average(data[data[:,3] >= 15, 1])
print(f"The result is {average_radio}")
```

##Get the sum of sales where the corresponding Newspaper is larger than the average of Newspaper

```python
average_newspaper = np.average(data[:,2])
sum_sales = np.sum(data[data[:,2] > average_newspaper, 3])
print(f"The result is {sum_sales}")
```

##Making the scores array

```python
average_sales = np.average(data[:,3])
scores = np.where(data[:,3] > average_sales, 'Good',
                  np.where(data[:, 3] < average_sales, 'Bad', 'Average'))

scores[7:10]
```

##Making the scores2 array

```python
closest_index = np.argmin(np.abs(data[:,3] - average_sales))
closest_value = data[closest_index, 3]

scores_2 = np.where(data[:,3] > closest_value, 'Good',
                  np.where(data[:, 3] < closest_value, 'Bad', 'Average'))

scores[7:10]
```
